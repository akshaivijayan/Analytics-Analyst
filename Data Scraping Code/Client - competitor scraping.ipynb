# Event Queen

```python
import requests
import collections
import numpy as np
import collections
import numpy as np
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
req = requests.get("https://rsgevents.ca/about/")
req1 = requests.get("https://vestievents.com/")
req2 = requests.get("https://eventqueen.ca/services")
soup= BeautifulSoup(req.content,"html.parser")
soup1 = BeautifulSoup(req1.content,"html.parser")
soup2 = BeautifulSoup(req2.content,"html.parser")
#print(soup.prettify())
res=soup.title
#at=soup.find(id="providerInfo")
at2 = soup2.find(id = "d471dc90-7503-4aff-9829-c07a22e68810")


#print(at.get_text())
#d = d + at1.get_text()  + a + b +"From intimate gatherings to larger scale weddings and events, the Chic by Nicole team is committed to working with couples to bring their visions to life and ensure their events run flawlessly.  Named one of the Top Wedding Planners worldwide by VOGUE Magazine and Black planner to follow by Martha Stewart Weddings, we offer bespoke services and design proposals that are both romantic, timeless and approachable. We provide full service wedding management and styling services in Toronto, Napa, Niagara Falls, New York, Paris and Worldwide"
d= at2.get_text() 
print(d) 


from matplotlib import rcParams
from wordcloud import WordCloud, STOPWORDS
all_headlines=d
stopwords = STOPWORDS
wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_words=1000).generate(all_headlines)
rcParams['figure.figsize'] = 10, 20

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
filtered_words = [word for word in all_headlines.split() if word not in stopwords]
counted_words = collections.Counter(filtered_words)
words = []
counts = []
for letter, count in counted_words.most_common(10):
    words.append(letter)
    counts.append(count)
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10
plt.title('Top words in the headlines vs their count')
plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)
plt.show()
```
